```python
import os
import csv

dataset_dir = '/teamspace/studios/this_studio/Celebrity Faces Dataset'

metadata_file = 'celebrity_metadata.csv'

def create_metadata(dataset_dir, metadata_file):
    with open(metadata_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['filename', 'celebrity_name'])  

        for root, dirs, files in os.walk(dataset_dir):
            if files:  # If there are files in the directory
                celebrity_name = os.path.basename(root)  # Get the celebrity name from the directory name
                for file in files:
                    if file.endswith('.jpg') or file.endswith('.jpeg') or file.endswith('.png'):
                        file_path = os.path.join(root, file)
                        writer.writerow([file_path, celebrity_name])

create_metadata(dataset_dir, metadata_file)
```
