```python
import tensorflow as tf
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input, decode_predictions
import cv2
import numpy as np

# Load the InceptionV3 model pre-trained on ImageNet data
model = InceptionV3(weights='imagenet')

# Function to preprocess and classify an image
def classify_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (299, 299))  # Resize image to expected model input size
    img = np.expand_dims(img, axis=0)  # Add batch dimension
    img = preprocess_input(img)  # Preprocess input as required by the model
    preds = model.predict(img)
    label = decode_predictions(preds, top=1)[0][0]  # Get the top prediction
    return label[1], label[2]  # Return the label and confidence

# Example usage
image_path = '/teamspace/studios/this_studio/Celebrity Faces Dataset/Scarlett Johansson/002_ea6e259d.jpg'
label, confidence = classify_image(image_path)
print(f'Label: {label}, Confidence: {confidence}')
```

```python
import tkinter as tk
from tkinter import filedialog, messagebox
import cv2
import numpy as np

class ImageBlurApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Image Blur Toggle")

        self.image_path = None
        self.image = None
        self.blurred_image = None
        self.is_blurred = True

        self.canvas = tk.Canvas(self.root, width=400, height=400)
        self.canvas.pack()

        self.load_button = tk.Button(self.root, text="Load Image", command=self.load_image)
        self.load_button.pack(pady=10)

        self.blur_button = tk.Button(self.root, text="Toggle Blur", command=self.toggle_blur)
        self.blur_button.pack(pady=10)

    def load_image(self):
        self.image_path = filedialog.askopenfilename(initialdir="./", title="Select Image",
                                                     filetypes=(("JPEG files", "*.jpg"), ("PNG files", "*.png"),
                                                                ("All files", "*.*")))
        if self.image_path:
            self.image = cv2.imread(self.image_path)
            self.show_image()

    def show_image(self):
        if self.image is not None:
            self.blurred_image = cv2.GaussianBlur(self.image, (15, 15), 0)
            self.show_blurred_image()

    def show_blurred_image(self):
        if self.is_blurred:
            img = cv2.cvtColor(self.blurred_image, cv2.COLOR_BGR2RGB)
        else:
            img = cv2.cvtColor(self.image, cv2.COLOR_BGR2RGB)

        img = cv2.resize(img, (400, 400))
        img = np.rot90(img)

        self.photo = tk.PhotoImage(data=img.tobytes())
        self.canvas.create_image(0, 0, anchor=tk.NW, image=self.photo)

    def toggle_blur(self):
        if self.image is None:
            messagebox.showwarning("No Image", "Please load an image first!")
            return

        self.is_blurred = not self.is_blurred
        self.show_blurred_image()


if __name__ == "__main__":
    root = tk.Tk()
    app = ImageBlurApp(root)
    root.mainloop()
```
